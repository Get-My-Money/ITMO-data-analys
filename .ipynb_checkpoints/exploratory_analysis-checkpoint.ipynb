### Активность за последний час

```python
# Создание новых признаков
main_df['last_hour_transactions'] = main_df['last_hour_activity'].apply(lambda x: x['num_transactions'])
main_df['last_hour_amount'] = main_df['last_hour_activity'].apply(lambda x: x['total_amount'])

# Разделение данных
fraud_data = main_df[main_df['is_fraud'] == 1]
legit_data = main_df[main_df['is_fraud'] == 0]

# Настройка графиков
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharex=True, sharey=True)

# График для мошеннических транзакций
sns.scatterplot(
    data=fraud_data,
    x='last_hour_transactions',
    y='last_hour_amount',
    ax=axes[0],
    alpha=0.7
)
axes[0].set_title('Мошеннические транзакции')
axes[0].set_xlabel('Количество транзакций за час')
axes[0].set_ylabel('Общая сумма за час')

# График для легитимных транзакций
sns.scatterplot(
    data=legit_data,
    x='last_hour_transactions',
    y='last_hour_amount',
    ax=axes[1],
    alpha=0.7
)
axes[1].set_title('Легитимные транзакции')
axes[1].set_xlabel('Количество транзакций за час')
axes[1].set_ylabel('Общая сумма за час')

plt.tight_layout()
plt.show()
```

Сильных различий между графиками визуально не обнаружено, кроме кол-ва операций, мошенеческих значительно меньше.

## Продуктовые гипотезы

### Гипотеза 1 - У клиентов с премиальными картами (Gold, Platinum) высокая доля легитимных транзакций за границей — значит, is_outside_home_country для них нужно интерпретировать мягче.

```python
premium_cards = ['Gold Credit', 'Platinum Credit', 'Platinum Debit']
main_df['is_premium'] = main_df['card_type'].isin(premium_cards)

result = main_df.groupby(['is_premium', 'is_outside_home_country'])['is_fraud'].agg(['mean', 'count'])
print(result)
```

Вывод: Независимо от типа карты (premium или нет), транзакции за пределами домашней страны имеют примерно одинаковую и очень высокую долю мошенничества (~56.7–56.9%), поэтому даже для таких клиентов не стоит ослабленять правила. Однако is_outside_home_country — один из самых сильных предикторов мошенничества, на который стоит обращать внимание.

Ценность для организации: Несмотря на то, что премиальные клиенты чаще совершают легитимные поездки, риск мошенничества за границей у них не ниже, чем у других. Это означает, что политика безопасности должна быть единообразной: нельзя снижать уровень проверок по типу карты. Вместо этого можно использовать повышенную активность в сочетании с геолокацией для запуска адаптивной аутентификации , сохраняя баланс между безопасностью и удобством.

### Гипотеза 2 - Мошеннические транзакции чаще происходят ночью (00:00–06:00 по часовому поясу страны), особенно в высокорисковых категориях.

```python
# Извлекаем час в часовом поясе страны (упрощённо — по timestamp)
main_df['hour'] = main_df['timestamp'].dt.hour

# Высокорисковые категории
high_risk_categories = ['gaming', 'travel', 'digital goods', 'entertainment']
main_df['is_high_risk_category'] = main_df['vendor_type'].str.lower().isin(high_risk_categories)

# Фильтруем ночные часы
night_mask = (main_df['hour'] >= 0) & (main_df['hour'] < 6)
night_high_risk = main_df[night_mask & main_df['is_high_risk_category']]

fraud_night = night_high_risk['is_fraud'].mean()
fraud_day = main_df[~night_mask & main_df['is_high_risk_category']]['is_fraud'].mean()

print(f"Fraud rate ночью: {fraud_night:.2%}")
print(f"Fraud rate днём: {fraud_day:.2%}")
```

Вывод: Доля мошенничества в ночные часы (00:00–06:00) составляет 50.15%, что значительно выше, чем днём (12.48%), что указывает на систематическую связь мошеннической активности со временем суток и подтверждает необходимость усиления проверок в этот период.

Ценность для организации: Высокая концентрация мошенничества в ночные часы позволяет оптимизировать нагрузку на систему мониторинга, перераспределив ресурсы в «пиковое» время. Организация может внедрить динамическое повышение уровня верификации ночью — например, обязательный 2FA для транзакций в игровых и цифровых сервисах, — что повысит эффективность детекции без увеличения ложных срабатываний днём.

### Гипотеза 3 - Введение функции «Я в поездке» в мобильном приложении снизит количество ложных срабатываний и chargebacks.

Реализация: Для проверки гипотезы предлагается внедрить в мобильное приложение функцию «Я в поездке», где пользователь может заранее указать страну и сроки пребывания за границей. Эта информация будет интегрирована в антифрод-систему, временно снижая вес признака is_outside_home_country и заменяя автоматическую блокировку транзакций на мягкие меры — такие как push-уведомление с подтверждением операции. Это позволит дифференцировать легитимные поездки от подозрительной активности.

Вывод (потенциальный): Если гипотеза подтвердится, можно будет сделать вывод, что проактивное поведение пользователя (самообъявление о поездке) является надёжным сигналом для снижения уровня риска. Это позволит системе перейти от универсальных правил к персонализированному подходу, уменьшив количество ложных срабатываний без потери безопасности.

Ценность для организации (потенциальная): Организация получит возможность повысить удовлетворённость клиентов, особенно среди часто путешествующих и премиальных пользователей, снизив число жалоб и обращений в поддержку из-за блокировок карт. Одновременно это укрепит доверие к платёжной системе и снизит операционные издержки, связанные с обработкой chargebacks и верификациями, сохраняя высокий уровень защиты от мошенничества.

## Технические гипотезы

### Гипотеза 1 - Мошенничество чаще происходит в странах с высоким уровнем финансовой уязвимости, что отражается в комбинации валюты, частоты транзакций и поведенческих паттернов — даже при сопоставимых суммах в USD.

```python
# Конвертируем все суммы в USD
main_df['date'] = main_df['timestamp'].dt.date
merged = main_df.merge(currency_df, left_on='date', right_on='date')

def to_usd(row):
    if row['currency'] == 'USD':
        return row['amount']
    try:
        return row['amount'] / row[row['currency']]
    except:
        return np.nan

merged['amount_usd'] = merged.apply(to_usd, axis=1)

# Выбираем страны с подозрительно высоким fraud rate
high_risk_countries = ['Nigeria', 'Russia', 'Brazil', 'Mexico', 'India']
merged['is_high_risk_country'] = merged['country'].isin(high_risk_countries)

# Сравниваем долю мошенничества при условии сопоставимых сумм
filtered = merged[merged['amount_usd'].between(10, 1000)]  # средние чеки
fraud_by_country = filtered.groupby('is_high_risk_country')['is_fraud'].agg(['mean', 'count'])
print(fraud_by_country)
```

Вывод: После перевода всех транзакций в USD и фильтрации по среднему чеку (10–1000 USD) было установлено, что доля мошенничества в таких странах, как Нигерия, Россия, Бразилия и Мексика, стабильно превышает 20%, в то время как в других регионах — около 2.5%. Это указывает, что высокий риск связан не с масштабом суммы или инфляцией, а с системными факторами: уровень киберпреступности, слабая защита онлайн-платежей, распространённость card-not-present мошенничества.

Польза для организации: Данные позволяют внедрить географический скоринг как устойчивый признак в ML-модели и правилах. Организация может настроить адаптивную безопасность: например, требовать 2FA для всех операций из high-risk стран, особенно при использовании веб-каналов или анонимных устройств. Это повысит эффективность защиты без ущерба для пользователей из низкорисковых регионов.

### Гипотеза 2 - Высокая активность за последний час (много транзакций, стран, мерчантов) сильно коррелирует с мошенничеством.

```python
# Распаковываем last_hour_activity
act = pd.json_normalize(main_df['last_hour_activity'])

# Применяем условия
high_risk_activity = (
    (act['num_transactions'] >= 5) &
    (act['unique_countries'] >= 2) &
    (act['unique_merchants'] >= 3)
)

fraud_rate_high = main_df[high_risk_activity]['is_fraud'].mean()
fraud_rate_total = main_df['is_fraud'].mean()

print(f"Доля fraud при высокой активности: {fraud_rate_high:.2%}")
print(f"Общая доля fraud: {fraud_rate_total:.2%}")
```

Вывод: Сам по себе всплеск активности (много транзакций, стран, мерчантов за час) не увеличивает долю мошенничества — она остаётся на уровне базового 20%. Однако при детализации выявлено, что транзакции из определённых стран (Нигерия, Россия, Бразилия, Мексика) имеют аномально высокую долю fraud — около 20%, против всего 2.5% у остальных. Это означает, что риск складывается не из одного фактора, а из их комбинации.

Польза для организации: Организация может использовать эту паттерн-зависимость для создания точечных правил безопасности. Например, запускать усиленную верификацию только при одновременном выполнении двух условий: высокая активность и операция из high-risk страны. Это снизит нагрузку на систему и минимизирует ложные срабатывания у низкорисковых пользователей, сохранив высокую чувствительность к реальным угрозам.

### Гипотеза 3 - Добавление признака "рост суммы относительно медианы клиента" улучшит точность модели детекции.

Реализация: Для проверки гипотезы нужно рассчитать индивидуальный поведенческий паттерн каждого клиента: на основе истории транзакций вычислить медианную сумму за последние 7 дней. Затем для каждой новой транзакции определенить отношение её суммы к этой медиане — amount_ratio. Транзакции с amount_ratio > 10 нужно выделить как потенциально аномальные. Далее сравнить долю мошенничества в группе с высоким ratio и в общей выборке.

Вывод (потенциальный): Если гипотеза подтвердится, можно ожидать, что транзакции с высоким значением amount_ratio (например, >10) будут демонстрировать значительно более высокую долю мошенничества по сравнению с остальными. Это будет означать, что резкие отклонения от привычного поведения клиента являются сильным индикатором мошеннической активности, даже если абсолютная сумма не является экстремальной.

Польза для организации (потенциальная): Организация сможет использовать этот признак для повышения чувствительности модели к индивидуальным аномалиям. Особенно эффективным он может оказаться для выявления мошенничества у клиентов с низким средним чеком, где даже умеренная по размеру транзакция может быть подозрительной. Это позволит сделать систему детекции более персонализированной, точной и менее зависимой от глобальных порогов, что в долгосрочной перспективе снизит количество как ложных срабатываний, так и пропущенных случаев fraud.
